```python
def importar_datos(): 
    import gspread
    import pandas as pd
    from oauth2client.service_account import ServiceAccountCredentials
    scope = ["https://spreadsheets.google.com/feeds","https://www.googleapis.com/auth/spreadsheets","https://www.googleapis.com/auth/drive.file","https://www.googleapis.com/auth/drive"]
    creds = ServiceAccountCredentials.from_json_keyfile_name("Aquí va la clave.json", scope)
    client = gspread.authorize(creds)

    sheet_inv = client.open("Inversión").sheet1 #guardo lo que hay en la hoja uno de la hoja de calculo "Inversión"
    sheet_gastos = client.open("Gastos").sheet1
    list_of_dicts = sheet_inv.get_all_records() #tomo una lista de diccionarios que haya ahí adentro
    list_of_dicts2 = sheet_gastos.get_all_records() 



    df_inversion1 = pd.DataFrame(list_of_dicts) #transformo a DF
    df_gastos = pd.DataFrame(list_of_dicts2)
    df_inversion = df_inversion1[["Monto invertido","monto total","intereses_diario"]] #tomo las columnas que me interesan


    list_columns_gastos = list(df_gastos.columns) 
    list_columns_inversion = list(df_inversion.columns)
    return df_gastos,df_inversion1,df_inversion
```

```python
from matplotlib import pyplot as plt
import pandas as pd

def calculo_intereses(ingresos,gastos,anios,inflacion_anual):
    
    inflacion_anual = inflacion_anual.get()
    
    anios = anios.get()
    
    ingresos = ingresos.get()
    
    gastos = gastos.get()
    
    infla_mensual = inflacion_anual/(12*100)
    
    intereses = []

    inv_neta = []
    
    for i in range(0,12 * anios): #esto es el rango de un año

        gastos = gastos * (1 + infla_mensual)

        inversión_neta = ingresos - gastos

        inv_neta.append(round(inversión_neta,3))

        if i == 0:

            inversión_relativa = inversión_neta 

        else:

            inversión_relativa = inversión_neta + intereses[i - 1] * (1 + 0.031)

        intereses.append(round(inversión_relativa ,3 ))

        #df = pd.DataFrame({"Rango": list(range(0,12)),"Intereses": intereses })
        #por si quiero armar un df
    return inv_neta, intereses, anios


```

```python
def ventana_grafica(lista_ingresos, lista_monto_tot,tiempo): 
    
    root2 = tkinter.Tk()
    root2.wm_title("Embedding in Tk")

    fig = Figure(figsize=(5, 4), dpi=100)
    tiempo = list(range(0,tiempo * 12))

    b = lista_monto_tot
    
    
    fig.add_subplot(111).plot(tiempo ,b  ) #acá tengo que meter mi función
    #inversión_relativa = inversión_neta + intereses[i - 1] * (1 + 0.031)
    
    canvas = FigureCanvasTkAgg(fig, master=root2)  # A tk.DrawingArea.
    
    canvas.draw()
    
    
    
    canvas.get_tk_widget().pack(side=tkinter.TOP, fill=tkinter.BOTH, expand=1)

    
    
    toolbar = NavigationToolbar2Tk(canvas, root2)
    toolbar.update()
    canvas.get_tk_widget().pack(side=tkinter.TOP, fill=tkinter.BOTH, expand=1)


    def on_key_press(event):
        print("you pressed {}".format(event.key))
        key_press_handler(event, canvas, toolbar)


    canvas.mpl_connect("key_press_event", on_key_press)


    def _quit():
        root2.quit()     # se para el "mainloop"
        root2.destroy()  # esto es necesario en la ventana
                        # para prevernir error de python FATAL : PyEval_RestoreThread: NULL tstate


    button = tkinter.Button(master=root2, text="Quit", command=_quit)
    button.pack(side=tkinter.BOTTOM)

    tkinter.mainloop()

    # si se pone el root.destroy() acá, va a causar un error si la ventana se
    # cierra con el administrador de ventanas.
```
